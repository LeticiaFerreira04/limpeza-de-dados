# Limpeza de dados da base de acidentes rodoviários nos EUA.

### Visualização e limpeza dos dados

### Link: https://www.kaggle.com/datasets/sobhanmoosavi/us-accidents

```python
# Importação da biblioteca

import pandas as pd
```

```python
# Leitura do arquivo

dataframe = pd.read_csv("/content/US_Accidents_Dec21_updated.csv", encoding="utf-8")
dataframe
```

```python
dataframe = dataframe.head(100000)
dataframe
```

```python
# Verificando a quantidade de linhas e colunas antes de remover as duplicatas

dataframe.shape
```

```python
# Removendo as linhas duplicadas

dataframe = dataframe.drop_duplicates()
dataframe.shape
```

```python
dataframe.info()
```

```python
# Trocando a nomenclatura

df_renomeado = dataframe.rename(columns= {"Number": "Street_Number"})
df_renomeado.info()
```

```python
# Convertendo as colunas 'Start_Time' e 'End_ Time para o tipo data

df_renomeado.Start_Time = pd.to_datetime(df_renomeado.Start_Time)
df_renomeado.End_Time = pd.to_datetime(df_renomeado.End_Time)
```

```python
# Visualização

df_renomeado.info()
```

```python
# Verificação de valores nulos

df_renomeado.isnull().sum()
```

```python
# Substituindo os valores nulos

df_renomeado["End_Lat"] = df_renomeado["End_Lat"].fillna("N/A")
df_renomeado["End_Lng"] = df_renomeado["End_Lng"].fillna("N/A")
df_renomeado["Street_Number"] = df_renomeado["Street_Number"].fillna("N/A")
df_renomeado["City"] = df_renomeado["City"].fillna("N/A")
df_renomeado["Sunrise_Sunset"] = df_renomeado["Sunrise_Sunset"].fillna("N/A")
df_renomeado["Civil_Twilight"] = df_renomeado["Civil_Twilight"].fillna("N/A")
df_renomeado["Nautical_Twilight"] = df_renomeado["Nautical_Twilight"].fillna("N/A")
df_renomeado["Astronomical_Twilight"] = df_renomeado["Astronomical_Twilight"].fillna("N/A")
df_renomeado["Timezone"] = df_renomeado["Timezone"].fillna("N/A")
df_renomeado["Airport_Code"] = df_renomeado["Airport_Code"].fillna("N/A")
df_renomeado["Weather_Timestamp"] = df_renomeado["Weather_Timestamp"].fillna("N/A")
df_renomeado["Wind_Direction"] = df_renomeado["Wind_Direction"].fillna("N/A")
df_renomeado["Weather_Condition"] = df_renomeado["Weather_Condition"].fillna("N/A")

df_renomeado["Zipcode"] = df_renomeado["Zipcode"].fillna(99999)

df_renomeado["Temperature(F)"] = df_renomeado["Temperature(F)"].fillna(999)
df_renomeado["Wind_Chill(F)"] = df_renomeado["Wind_Chill(F)"].fillna(999)

df_renomeado["Humidity(%)"] = df_renomeado["Humidity(%)"].fillna(0)
df_renomeado["Pressure(in)"] = df_renomeado["Pressure(in)"].fillna(0)
df_renomeado["Visibility(mi)"] = df_renomeado["Visibility(mi)"].fillna(0)
df_renomeado["Wind_Speed(mph)"] = df_renomeado["Wind_Speed(mph)"].fillna(0)
df_renomeado["Precipitation(in)"] = df_renomeado["Precipitation(in)"].fillna(0)
```

```python
# Verificando se todos os campos foram preenchidos

df_renomeado.isnull().sum()
```
